```python
import pandas as pd
import numpy as np
sentences =['الطبيعة الجميلة الخلابه و الطقس في الشتاء',
           'في الصيف و الربيع الطقس جميل',
           ' الطقس الربيع هنا يتكاثر ثم ينام وحيداً']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec=CountVectorizer()
X= vec.fit_transform(sentences)
bag_of_words = pd.DataFrame(X.toarray() , columns=vec.get_feature_names())
bag_of_words
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec= TfidfVectorizer()
X = vec.fit_transform(sentences )
bag_of_words = pd.DataFrame(X.toarray() , columns=vec.get_feature_names())
np.round(bag_of_words ,2)
```
